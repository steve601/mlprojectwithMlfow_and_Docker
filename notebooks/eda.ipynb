```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### let's load the clean data

```python
df = pd.read_csv('C:/Users/odhia/OneDrive/Desktop/datascienceproject/data/clean_dataa.csv')
df.head()
```

### distribution plots for the columns

```python
cols = ['pregnancies', 'glucose', 'bloodpressure', 'skinthickness', 'insulin',
       'bmi', 'diabetespedigreefunction', 'age']
for i in cols:
    sns.histplot(x=df[i])
    plt.title(f'Histplot  (for {i})')
    plt.ylabel(f'{i}')
    plt.show()
```

```python
for i in cols:
    sns.boxplot(x=df['outcome'], y=df[i], data=df)
    plt.title(f'BoxPlot of Outcome (for {i})')
    plt.ylabel(f'{i}')
    plt.show()
```

### plot for bloodpressure against glucose,with respect to outcome

```python
sns.scatterplot(x = df.glucose,y = df.bloodpressure,hue = df.outcome)
# people with diabetes have between 60 - 100 for both glucose and blood sugar level
```

### plot for age aginst bmi,with respect to outcome

```python
sns.scatterplot(x = df.bmi,y = df.age,hue = df.outcome)
# most people with diabetes are of below 30 years and have low bmi
```
